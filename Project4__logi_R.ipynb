```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
data = pd.read_csv("/content/LoanApprovalPrediction.csv")
```

```python
data.head()
```

```python
data.dropna()
```

```python
data.replace({'Married':{'Yes':1,'No':0},'Gender':{'Male':1,'Female':0},'Self_Employed':{'Yes':1,'No':0},'Property_Area':{'Rural':0,'Urban':1,'Semiurban':2},'Loan_Status':{'Y':1,'N':0}},inplace=True)
```

```python
data.head()
```

```python
data.replace({'Education':{'Graduate':1,'Not Graduate':0}},inplace=True)
```

```python
data.drop(['Loan_ID'],axis=1,inplace=True)
```

```python
for col in data.columns:
    data[col]=data[col].fillna(data[col].mean())
```

```python
x=data.drop(['Loan_Status'],axis=1)
y=data['Loan_Status']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.70)
```

```python
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
x_train_pred = model.predict(x_train)
```

```python
Acc_train = accuracy_score(y_train,x_train_pred)
print(Acc_train)
```

```python
x_test_pred = model.predict(x_test)
Acc_test = accuracy_score(y_test,x_test_pred)
print(Acc_test)
```
