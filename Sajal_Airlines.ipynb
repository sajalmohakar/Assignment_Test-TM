```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import svm
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
```

```python
airline = pd.read_csv("/content/airline.csv")
```

```python
airline.head()
```

```python
airline.tail()
```

```python
airline.isnull().sum()
```

```python
airline.shape
```

```python
airline["satisfaction"].value_counts()
```

```python
airline.head()
```

```python
airline.replace({"satisfaction":{'satisfied':1,'neutral or dissatisfied':0}},inplace=True)
```

1 = Statisified
0 = neutral or dissatisfied

```python
x  = airline.drop(['satisfaction'],axis=1)
y= airline['satisfaction']
```

```python
y.head()
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(x)
```

```python
scaler.transform(x)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.5)
```

```python
x_train.shape,x_test.shape
```

```python
model = LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
train_data_pred = model.predict(x_train)
```

```python
Train_Accuracy_score = accuracy_score(y_train,train_data_pred)
```

```python
print(Train_Accuracy_score)
```

```python
test_data_pred = model.predict(x_test)
```

```python
Test_Accuracy_score = accuracy_score(y_test,test_data_pred)
```

```python
print(Test_Accuracy_score)
```

```python
# Prediction model
input = (54,468,1,4,1,4,4,1,4,4,4,5,5,3,4,4,0,0.0,0,1,0,1,0,1,0)
input_into_numpy = np.asarray(input)
input_reshaped = input_into_numpy.reshape(1,-1)
std_data = scaler.transform(input_reshaped)
prediction = model.predict(std_data)
print(prediction)
```
